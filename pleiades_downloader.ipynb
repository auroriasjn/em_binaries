```python
from astroquery.gaia import Gaia
from downloader import GaiaAggregator
from utils import create_gaia_query

import pandas as pd

%load_ext autoreload
%autoreload 2
```

```python
# Aggregate weirdly formatted files
analyzer = GaiaAggregator(file_names=["data/clustercat.dat", "data/escapecat.dat"])
data = analyzer.get_data()

# Get the source ids
source_ids = data['Gaia_EDR3_Source_ID'].tolist()
source_ids
```

```python
adql_query = create_gaia_query(source_ids)

job = Gaia.launch_job_async(adql_query)
results = job.get_results()
results.write('gaia.csv', format='csv', overwrite=True)
```
